##1. Insertion and Time

````python
def contains(bag, e) :
    return e in bag

def insert(bag, e) :
    bag.append(e)

def insert_first(bag, e) :
    bag.insert(0, e) #0의 앞에 e를 넣는다. ([e])[0][1][2][3][4]

def remove(bag, e) :
    bag.remove(e) #e를 제거

def count(bag):
    return len(bag) #bag이 몇칸인지 출력
````

````python
myBag = [] #myBag, 리스트 초기화
insert(myBag, '휴대폰') # def insert(bag, e)와 연동
insert(myBag, '지갑')
insert(myBag, '손수건')
insert(myBag, '빗')
insert(myBag, '자료구조')
insert(myBag, '야구공')
insert_first(myBag, '1번')
print('가방속의 물건:', myBag)

insert(myBag, '빗')
remove(myBag, '손수건')
insert_first(myBag, '0번')
print('가방속의 물건:', myBag)
count(myBag)
````


## 2. What Is Algorithm Analysis?

* 1~n까지의 Sum을 구하는 함수

````python
def sum_of_n(n):#T=1+N+1=N+2 / O(N)
    the_sum = 0
    for i in range(1,n+1): #range(1,5)=[1,2,3,4] / range(5)=[0,1,2,3,4]
        the_sum = the_sum + i
    return the_sum
````

````python
print(sum_of_n(10))
````

* 같은 기능을 수행하는 함수이지만 이해하기 어려운 변수명, 추가적인 할당

````python
def foo(tom):
    fred = 0
    for bill in range(1, tom+1): #1~tom까지
        barney = bill
        fred = fred + barney
    return fred
````

````python
print(foo(10))
````

````python
import time

def sum_of_n_2(n):
    start = time.time()
    the_sum = 0

    for i in range(1, n+1):
        the_sum = the_sum + i
    end = time.time()

    return the_sum, end - start
````

````python
for i in range(5):
    print("Sum is %d required %10.7f seconds" % sum_of_n_2(10000)) #1-10000까지 더하는 함수
````

````python
for i in range(5):
    print("Sum is %d required %10.7f seconds" % sum_of_n_2(1000000)) #1-100000까지
````

````python
def sum_of_n_3(n):#T=1 / O(1)
    return (n * (n + 1)) / 2
````

````python
print(sum_of_n_3(10))
````

````python
def sum_of_n_3(n):
    start = time.time()
    the_sum = (n * (n + 1)) / 2 #for문에 비해 시간이 단축됨
    end = time.time()

    return the_sum, end - start
````

````python
for i in range(5):
    print("Sum is %d required %10.7f seconds" % sum_of_n_3(10000 * (10 ** (i+1)))) #10**(i+1)=10의 (i+1)제곱
````

###p.23 Example

````python
def sum_quad(n): #3제곱 계산
    partialSum = 0
    for i in range(n):#T=1+N+1=N+2 / O(N)
        partialSum += i*i*i
    return partialSum
````

````python
sum_quad(5)
````

### Rule 3

````python
N = 100
M = 40
k = 0
for i in range(N):
    for j in range(N): #O(N^2)
        k = i + j
for i in range(M): #O(M)
    k += i #O(N^2)+O(M)=O(N^2)

print(k)
````

### p.31 순환 Recursive

* factorial

````python
def factorial(n): #o(n)
    if n==1:
        return 1
    else:
        return n*factorial(n-1)
````

````python
print(factorial(10))
````

````python
def factorial2(n): #o(n)
    result = 1
    for k in range(n,0,-1):
        result = result * k
    return result
````

````python
print(factorial2(10))
````

* p.36 피보나치 수열

````python
import time
````

````python
def fib(n): #피보나치의 경우 recursive가 매우 비효율적
    if n == 0:
        return 0
    elif n == 1 or n == 2:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)#O(2^N)

n = int(input("숫자를 입력하세요."))

start = time.time()
print(fib(n))
end = time.time()
````

````python
start
````

````python
end
````

````python
print("running time : %10.7f seconds" % (end - start))
````

````python
def fib_iter(n): #효율적
    if (n < 2):
        return n
    last = 0
    current = 1
    for i in range(2, n+1):#O(N)
        tmp = current
        current += last
        last = tmp
    return current

n = int(input("숫자를 입력하세요."))
start = time.time()
print(fib_iter(n))
end = time.time()
````

````python
start
````

````python
end
````

````python
print("running time : %10.7f seconds" % (end - start))
````

````python
#matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def exponential(n):
  return 3**n
def cubic(n):
  return n**3
def quadratic(n):
  return n**2
def loglinear(n):
  return n*np.log(n) #nlog(n)
def linear(n):
  return n
def logarithmic(n):
  return np.log(n)
def constant(n):
  return np.ones(len(n)) #길이
````

````python
n=np.arange(0, 5) #[0, 1, 2, 3, 4]

plt.plot(n, exponential(n), c='red') #[3^0, 3^1, 3^2, 3^3, 3^4]
plt.plot(n,cubic(n), c='orange')
plt.plot(n, quadratic(n), c='yellow')
plt.plot(n,loglinear(n), c='green')
plt.plot(n,linear(n), c='blue')
plt.plot(n,logarithmic(n), c='cyan')
plt.plot(n,constant(n), c='magenta')

plt.xlabel('Size of input(n)', fontsize=16)
plt.ylabel('Number of operations', fontsize=16)
````
